This imports all of the functions necessary and creates the connectable

```python
import sqlite3
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn import linear_model


database_service = "sqlite"
database = "jupiter.db"
connectable = f"{database_service}:///{database}"
```

This prints out all of the variables in the jupiter.db database 

```python
connection = sqlite3.connect('jupiter.db')

cursor = connection.cursor()

columns_query = "PRAGMA table_info(moons);" #Query to retrieve information about the columns in the moons table
cursor.execute(columns_query)
columns = cursor.fetchall() #Fetch all the columns' information from the executed query

print("Column names for the 'moons' table:")
for column in columns:
    print(column[1]) #for loop which runs through the titles in the database table and prints them out

cursor.close()
connection.close()
```

This imports my jupiter python code and uses the Moons class to work out the stats and correlations between variables

```python
from jupiter import Moons #imports python code

jupiter_moons = Moons('jupiter.db') #inputs the database into the class 

correlations = jupiter_moons.correlations()
print("Correlation Statistics:")
display(correlations) #displays all correlations between variables
sns.heatmap(data=correlations, annot=True) #displays correlations in a heatmap format
plt.show()


summary_stats = jupiter_moons.stats() #produces statistics about the variables
print("Summary Statistics:")
display(summary_stats)

jupiter_moons.plot_hist()

jupiter_moons.group_plot()

io_data = jupiter_moons.moon('Io') #displays information about Io as an example, any other moon inputed would work
print("\n \n \nData for Io:")
display(io_data)

print("\n \n \nHere are some main statistics from the database: \n")

mooncount = (summary_stats["period_days"]["count"])
print(f"There are {mooncount} moons in the database \n") #shows the amount of moons in the database

positivecorr = correlations[(correlations > 0.7) & (correlations != 1)  ] #filters for correlations between 0.7 and 1 (very positive)
print("Here are the positive correlations between the variables (coefficient >0.7):")
display(positivecorr)
countp1 = positivecorr.count().sum() #counts the numbers of positive correlations
countp2 = (countp1)/2
print(f"There are {countp2} positive correlations between variables in the dataframe \n")
print("These are between: \nPeriod days and distance (km) \nPeriod days and inclination degrees \nPeriod days and magnitude \nPeriod days and ecc \nDistance (km) and magnitude \nDistance (km) and ecc \nDistance (km) and inclination degrees \nRadius (km) and mass (kg) \n")

negativecorr = correlations[(correlations < -0.7)] #same again but with negative correlations
print("Here are the negative correlations between the variables (coefficient <-0.7):")
display(negativecorr)
countn1 = negativecorr.count().sum()
countn2 = (countn1)/2
print(f"There are {countn2} negative correlations between variables in the dataframe\n")
print("These are between: \nMagnitude and radius (km) \nMagnitude and mass (kg)\n")

print("These are how many pieces of data are missing for each variable:")
for column in columns:
    if column[1] != 'group': #loops through the variables and counts and displays how many data points are missing
        variables = (column[1])
        query = f"SELECT COUNT(*) FROM moons WHERE {variables} IS NULL;"
        missing_count = pd.read_sql(query, connectable)
        count = missing_count.iloc[0, 0]
        print(f"Missing count for variable {variables}: {count}")

jupiter_moons.linear_regression_model() #calls the linear regression model function
```
